```python
import csv

input_file = "knesset_sentences.csv"
output_file = "knesset_filtered_sentences.csv"
text_column = "sentence_text"   # name of the column to check

hebrew_keywords = [
    "פליטת",
    "ג'נין",
    "ירדן",
    "פחמן",
    "חממה",
    "גזים",
    "כושים",
    "כלי רכב",
    "שועפאט",
    "מכונית",
    "דיזל",
    "בנזין",
    "פליטות"
]
# --------------

with open(input_file, newline="", encoding="utf-8") as infile, \
     open(output_file, "w", newline="", encoding="utf-8") as outfile:

    reader = csv.DictReader(infile)
    writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
    writer.writeheader()

    for row in reader:
        text = row[text_column]

        # Check if any forbidden Hebrew keyword appears
        if not any(keyword in text for keyword in hebrew_keywords):
            writer.writerow(row)
```
